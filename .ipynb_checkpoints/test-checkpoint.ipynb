```python
import cv2
from sahi.predict import get_prediction
from sahi import AutoDetectionModel
from PIL import Image
import matplotlib.pyplot as plt
import os

# Initialize the YOLOv8 model
detection_model = AutoDetectionModel.from_pretrained(
    model_type="yolov8",
    model_path='best_large_22img.pt',
    confidence_threshold=0.3,
    device="cpu",  # or 'cuda:0'
)

# Load the image to get its dimensions
image_path = "temp_image.jpg"
image = Image.open(image_path)
width, height = image.size
full_shape = [height, width]

# Get predictions from the SAHI model
result = get_prediction(image_path, detection_model, full_shape=full_shape)

bounding_boxes = []

# Iterate through all object predictions and get bounding boxes
for prediction in result.object_prediction_list[:2000]:  # Limit to 2000 detections
    bbox = prediction.bbox.to_xyxy()  # Get the bbox coordinates (x_min, y_min, x_max, y_max)
    bounding_boxes.append(bbox)  # Append the bounding box to the list

print(f"Number of bounding boxes: {len(bounding_boxes)}")

# Load the image with OpenCV
image = cv2.imread(image_path)

# Convert BGR to RGB (since OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Iterate over each bounding box and draw it on the image
for bbox in bounding_boxes:
    x_min, y_min, x_max, y_max = map(int, bbox)
    # Draw rectangle
    cv2.rectangle(image_rgb, (x_min, y_min), (x_max, y_max), color=(0, 255, 0), thickness=2)
    # Add label (optional)
    # cv2.putText(image_rgb, "object", (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# Show the image using Matplotlib
plt.figure(figsize=(15, 15))
plt.imshow(image_rgb)
plt.axis("off")
plt.show()

# Convert back to BGR format to save the image correctly with OpenCV
image_bgr = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR)

# Define the save path
output_dir = 'output_POC_2'
os.makedirs(output_dir, exist_ok=True)
image_name = os.path.basename(image_path)
output_path = os.path.join(output_dir, image_name)

# Save the image
cv2.imwrite(output_path, image_bgr)

print(f"Image saved at: {output_path}")
```
